# ETL

En esta primera parte voy a estar  leyendo un archivo llamado "steam_games.json" que contiene datos en formato JSON y convirtiéndolos en un DataFrame de pandas.
El código lee el archivo "steam_games.json" línea por línea, interpreta cada línea como un diccionario (ya que está en formato JSON) y almacena todos los diccionarios en un DataFrame de pandas para su posterior manipulación y análisis.

```python
import ast
import pandas as pd

rows = []
with open("steam_games.json") as f:
    for line in f.readlines():
        rows.append(ast.literal_eval(line))

df = pd.DataFrame(rows)
```

```python
# Imprimo el DataFrame para verificar que se haya leído correctamente.
df
```

Considero que como no se especificó un procesamiento de ETL, consideré importante borrar los registros con valores nulos en las columnas que considero indispensables, como lo son AppName y ID.

```python
# Contar el número de registros con valores nulos en las columnas "app_name" e "id"
registros_con_nulos = df[df["app_name"].isna() | df["id"].isna()].shape[0]

# Imprimir el resultado
print(f"Se eliminarían {registros_con_nulos} registros con valores nulos en 'app_name' o 'id'.")
```

```python
# Modificar el DataFrame original eliminando los registros con valores nulos en "app_name" e "id"
df.dropna(subset=["app_name", "id"], inplace=True)
```
